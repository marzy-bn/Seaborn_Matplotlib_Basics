```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

#create DataFrame
my_df = pd.DataFrame(randn(50,4), columns=['Mon','Tues','Wed','Thurs'])
my_df
```

```python
#scatter plot
my_df.plot(kind="scatter",x="Mon",y="Tues", c="Wed")
```

```python
my_df.plot(kind="scatter",x="Mon",y="Tues", c="Wed", cmap='magma')
```

```python
#By size
my_df.plot(kind="scatter",x="Mon",y="Tues", s=my_df['Wed']*100)
```

```python
#By size easier to read with alpha
my_df.plot(kind="scatter",x="Mon",y="Tues", s=my_df['Wed']*100, alpha=0.4)
```

```python
#By title
my_df.plot(kind="scatter",x="Mon",y="Tues", s=my_df['Wed']*100, alpha=0.4, title="My Awesome ScatterPlot")
```
