```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

#create a quick dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon","Tues","Wed","Thur"])
my_df
```

```python
#area plot
my_df.plot(kind='area', stacked=False)
```

```python
#area plot
my_df.abs().plot(kind='area')
```

```python
#shading
my_df.abs().plot(kind='area', alpha=0.3)
```

```python
#Title
my_df.abs().plot(kind='area', alpha=0.5, title="My Awesome area plot")
```

```python
#legend
my_df.abs().plot(kind='area', alpha=0.5, title="My Awesome area plot", legend=False)
```

```python
#plot only one column
my_df['Mon'].abs().plot(kind='area', alpha=0.5, title="My Awesome area plot", legend=False)
```

```python
#plot 2 cols
my_df[['Mon','Thur']].abs().plot(kind='area', alpha=0.5, title="My Awesome area plot", legend=False)
```

```python
#get info on function
my_df[['Mon','Thur']].abs().plot(kind='area', alpha=0.5, title="My Awesome area plot", legend=False)
```

```python
#table
my_df[['Mon','Thur']].abs().plot(kind='area', alpha=0.5, title="My Awesome area plot", legend=False, table=True)
```
